```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

data=pd.read_csv('Data/data.csv',encoding = 'UTF-8')

data.replace([np.inf, -np.inf], np.nan, inplace=True)

##normalise data
data['2012'] = (data['2012']-data['2012'].min())/(data['2012'].max()-data['2012'].min())
data['2013'] = (data['2013']-data['2013'].min())/(data['2013'].max()-data['2013'].min())  
data['2014'] = (data['2014']-data['2014'].min())/(data['2014'].max()-data['2014'].min())  
data['2015'] = (data['2015']-data['2015'].min())/(data['2015'].max()-data['2015'].min())  
data['2016'] = (data['2016']-data['2016'].min())/(data['2016'].max()-data['2016'].min())  
data['2017'] = (data['2017']-data['2017'].min())/(data['2017'].max()-data['2017'].min())  
data['2018'] = (data['2018']-data['2018'].min())/(data['2018'].max()-data['2018'].min())  
data['2019'] = (data['2019']-data['2019'].min())/(data['2019'].max()-data['2019'].min())  
data['2020'] = (data['2020']-data['2020'].min())/(data['2020'].max()-data['2020'].min())  

data['LGA_crime'] = ''

parallel_coordinates(data[['2012','2013','2014','2015','2016','2017','2018','2019','2020','LGA_crime']],'LGA_crime')
plt.savefig('Outputs/parallel_coords.jpeg', dpi = 300)
plt.show()
```
