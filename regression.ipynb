```python
import pandas as pd

data = pd.read_csv('Data/Combined.csv', encoding = 'UTF-8')

# Transforming the dataframe

data = data[['LGA_crime','2011_crime','2012_crime','2013_crime','2014_crime','2015_crime','2016_crime','2017_crime','2018_crime','2019_crime','2020_crime','2011_population','2012_population','2013_population','2014_population','2015_population','2016_population','2017_population','2018_population','2019_population','2020_population']]

data_2011 = data[['LGA_crime','2011_crime','2011_population']]
data_2011.columns = ['LGA','CR','Pop']
data_2011.insert(0,'2011','1')
data_2011 = data_2011[['LGA','CR','Pop','2011']]

data_2012 = data[['LGA_crime','2012_crime','2012_population']]
data_2012.columns = ['LGA','CR','Pop']
data_2012.insert(0,'2012','1')
data_2012 = data_2012[['LGA','CR','Pop','2012']]

data_2013 = data[['LGA_crime','2013_crime','2013_population']]
data_2013.columns = ['LGA','CR','Pop']
data_2013.insert(0,'2013','1')
data_2013 = data_2013[['LGA','CR','Pop','2013']]

data_2014 = data[['LGA_crime','2014_crime','2014_population']]
data_2014.columns = ['LGA','CR','Pop']
data_2014.insert(0,'2014', '1')
data_2014 = data_2014[['LGA','CR','Pop','2014']]

data_2015 = data[['LGA_crime','2015_crime','2015_population']]
data_2015.columns = ['LGA','CR','Pop']
data_2015.insert(0,'2015','1')
data_2015 = data_2015[['LGA','CR','Pop','2015']]

data_2016 = data[['LGA_crime','2016_crime','2016_population']]
data_2016.columns = ['LGA','CR','Pop']
data_2016.insert(0,'2016','1')
data_2016 = data_2016[['LGA','CR','Pop','2016']]

data_2017 = data[['LGA_crime','2017_crime','2017_population']]
data_2017.columns = ['LGA','CR','Pop']
data_2017.insert(0,'2017','1')
data_2017 = data_2017[['LGA','CR','Pop','2017']]

data_2018 = data[['LGA_crime','2018_crime','2018_population']]
data_2018.columns = ['LGA','CR','Pop']
data_2018.insert(0,'2018','1')
data_2018 = data_2018[['LGA','CR','Pop','2018']]

data_2019 = data[['LGA_crime','2019_crime','2019_population']]
data_2019.columns = ['LGA','CR','Pop']
data_2019.insert(0,'2019','1')
data_2019 = data_2019[['LGA','CR','Pop','2019']]

data_2020 = data[['LGA_crime','2020_crime','2020_population']]
data_2020.columns = ['LGA','CR','Pop']
data_2020.insert(0,'2020','1')
data_2020 = data_2020[['LGA','CR','Pop','2020']]

frames = [data_2011, data_2012, data_2013, data_2014, data_2015, data_2016, data_2017, data_2018, data_2019, data_2020]

result = pd.concat(frames)
result = result.fillna(0)

result.to_csv(path_or_buf = 'Data/PreRegression.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('Data/PreRegression.csv', encoding = 'UTF-8')

data = data[['CR','Pop','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']]

data = data.replace(',','', regex=True)
data.apply(pd.to_numeric, errors='raise')

X = data[['Pop','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']]

y = data['CR']

regr = LinearRegression(fit_intercept='True')
regr.fit(X, y)

# Make predictions using the testing set
y_pred = regr.predict(X)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y, y_pred))
```
